```python
##TRANSFORMAÇÃO PARA RELAÇÃO: CADA PERSONAGEM DO JOGO ESTABELECE UMA LIGAÇÃO COM OUTRO QUANDO ESTÁ NO MESMO DECK
### IMPORTANTE NÃO HÁ DIFERENÇA NA RELAÇÃO ENTRE X-Bow_> Knight ou 	Knight -> X-Bow OU SEJA É UM GRAFO E NÃO DIGRAFO

todos_os_pares = []
for deck in df.itertuples():
  sequencia = list(deck[2::])
  todos_os_pares.extend(list(combinations(sorted(sequencia), 2)))

contagem = dict(Counter(todos_os_pares))
pares=pd.DataFrame(contagem.items(), columns=['Relação','Contagem'])
pares
```

## ESTUDO DA BIBLIOTECA NETWORKX - GRAFOS

```python

G = nx.Graph() ##definindo o grafo

for item in pares.itertuples(): ###vamos add as arestas, os edges.

  G.add_edge(item.Relação[0], item.Relação[1], weight=item.Contagem, dist=1.0/(item.Contagem))

##VER NÓS E ARESTAS
print(G.nodes())
print(G.edges())

## VER UM NÓ ESPECIFICO: EXEMPLO print(G['The Log'])
```

```python
# 3. MÉTRICAS DE CENTRALIDADE DA REDE

# PAGERANK: "O PRESTÍGIO"
# Não é só aparecer muito, é aparecer com outras cartas que também são importantes.
# Se o PageRank é alto, a carta é "VIP": ela está nos melhores decks do meta.
pr = nx.pagerank(G, weight='weight')
# EIGENVECTOR: Ideia parecida com pagerank
eigen = nx.eigenvector_centrality(G, weight='weight', max_iter=1000)

# CLOSENESS: "O CORINGA"
# Mede a distância da carta para todas as outras. Usamos 1/peso (dist) para
# que cartas com muita sinergia sejam consideradas "perto".
# Quem ganha aqui é a carta mais versátil, que encaixa em quase qualquer deck.
closeness = nx.closeness_centrality(G, distance='dist')

# BETWEENNESS: "A PONTE"
# Mostra cartas que ligam grupos que normalmente não se conversam (ex: liga o grupo de Balão ao de X-Besta).
# Se você tirar uma carta com Betweenness alto, os grupos ficam isolados e o meta trava.
between = nx.betweenness_centrality(G, weight='dist')



# DEGREE: "CONEXÕES"
# É apenas a contagem direta de com quantas cartas diferentes ela faz combo.
# Se o Zap tem Grau 50, ele faz par com 50 cartas diferentes na  base de dados.
degree = dict(G.degree())
```

```python
## Distribuição da REDE + Cluster

pos = nx.spring_layout(G, k=0.75, weight='weight', iterations=200, seed=42) ##usamos modelo spring layout, com seed para manter sempre mesmo padrão

### OBTEMOS AS COORDENADAS DOS NÓS COM BASE NO POS DISTRIBUIÇÃO
X_coords = np.array([pos[node] for node in G.nodes()])

##VAMOS USAR AS COORDENADAS PARA APLICAR K MEANS, MODELO DE CLUSTERIZAÇÃO.
## O KMEANS DIFERENTES DE OUTROS MODELOS DE CLUSTERIZAÇÃO PRECISA QUE PASSEMOS VALOR INICIAL DE CLUSTER, QUE PODEMOS MANEJAR MANUALMENTE.
### VAMOS USAR COMO AUXÍLIO SILHOUETTE PARA DETERMINAR MELHOR QUANTIDADE.

sil_scores = []
k_range = range(4, 11) # K DE 4 A 10 ESCOLHA NOSSA BASEADO NO CONHECIMENTO PRÉVIO DOS DADOS CLASH ROYALE
for k in k_range:
    km = KMeans(n_clusters=k, random_state=42, n_init=10)
    labels = km.fit_predict(X_coords)
    sil_scores.append(silhouette_score(X_coords, labels)) ##ADD OS SILHOUETTE A LISTA PARA SELECIONAR O MELHOR

best_k=k_range[sil_scores.index(max(sil_scores))]
print(f"Melhor K encontrado: {best_k}") ## quantida otima de clusters, não é uma regra rígida

kmeans_final = KMeans(n_clusters=best_k, random_state=42, n_init=10).fit(X_coords)
kmeans_final.labels_


clusters = {node: int(kmeans_final.labels_[i]) for i, node in enumerate(G.nodes())}
print(clusters)
```

## EXPORTAR MÉTRICAS E ANALISES

```python



# 5. EXPORTAÇÃO DE DADOS E INSIGHTS NO CONSOLE
metrics_df = pd.DataFrame({
    'PageRank': pd.Series(pr),
    'Closeness': pd.Series(closeness),
    'Betweenness': pd.Series(between),
    'Eigenvector': pd.Series(eigen),
    'Grau': pd.Series(degree),
    'Cluster': pd.Series(clusters)
})

print("\n" + "="*70)
print(f"📊 RELATÓRIO ANALÍTICO FINAL (Melhor K encontrado: {best_k})")
print("="*70)

# Rankings Globais
for m in ['PageRank', 'Betweenness', 'Closeness']:
    print(f"\n🔥 TOP 5 - {m.upper()}:")
    print(metrics_df[m].sort_values(ascending=False).head(5))

print("\n" + "-"*70)
print("📂 ANÁLISE DE CADA ARQUÉTIPO (CLUSTERS)")
print("-"*70)

for i in range(best_k):
    cluster_cards = metrics_df[metrics_df['Cluster'] == i]
    if not cluster_cards.empty:
        core_card = cluster_cards['PageRank'].idxmax()
        names = cluster_cards.index.tolist()
        print(f"Grupo {i} | Core: {core_card}")
        print(f"   Cartas: {', '.join(names)}")


print("="*70)
print("💡 INSIGHTS ESTRATÉGICOS")
print("="*70)
print(f"1. A CARTA PONTE: '{metrics_df['Betweenness'].idxmax()}' é o maior elo entre diferentes decks.")
print(f"2. O CORINGA: '{metrics_df['Closeness'].idxmax()}' é a carta com maior versatilidade geométrica.")
print(f"3. ESTILO DOMINANTE: O Cluster {metrics_df['Cluster'].value_counts().idxmax()} é o maior grupo de cartas")
```

1. Log Bait  - Grupo 0
Por que esse nome? Você tem Goblin Barrel, Princess, Goblin Gang e Inferno Tower.

A lógica: Esse é o conjunto clássico de cartas que tentam forçar o oponente a gastar o "Tronco" (Log) para que o Barril de Goblins possa dar dano. O Rocket apareceu como Core porque ele é a condição de vitória pesada desse estilo de deck.

2. Beatdown - Grupo 1
Por que esse nome? Temos Royal Giant, Mega Knight, Elite Barbarians e Three Musketeers.

A lógica: São cartas pesadas que focam em "atropelar" o adversário criando um push gigante. A Fireball é o Core porque, nesses decks pesados, ela é essencial para limpar o caminho.

3. Ciclo Rápido - Grupo 2
Por que esse nome? Estão aqui o The Log, Hog Rider (Corredor), Ice Spirit, Skeletons e Knight.

A lógica: São cartas baratas (2.6 ou similares). O objetivo é rodar o deck o mais rápido possível para jogar o Corredor várias vezes. O The Log é o Core porque é a carta mais onipresente nesse estilo de jogo.

4. Champions/Lendárias  - Grupo 3
Por que esse nome? Aparecem Tornado, Electro Giant, Golden Knight e Graveyard.

A lógica: São cartas que exigem muita técnica e geralmente giram em torno de combos específicos com cartas lendárias ou campeões. O Tornado é o Core porque ele "puxa" a estratégia desses decks (como o combo com Bebê Dragão ou Gigante Elétrico).

5. Tank/Suporte  - Grupo 4
Por que esse nome? Estão aqui PEKKA, Zap, Arrows, Wizard e Executioner.

A lógica: É o grupo das "ferramentas". São cartas usadas para defender e dar suporte a tanques. O Barbarian Barrel é o Core aqui porque é a unidade de utilidade que mais faz o "trabalho sujo" de defesa nesse grupo.

6. Construções/Defesa  - Grupo 5
Por que esse nome? Temos X-Bow (X-Besta), Tesla, Archer Queen e Earthquake.

A lógica: É o famoso deck de "Pedreiro" (Siege). O objetivo é montar uma defesa impenetrável e atacar do seu próprio lado do campo. A Archer Queen é o Core porque ela é a unidade de defesa/contra-ataque mais forte dessa categoria.

##VISUALIZAÇÃO GRAFO

```python

#VISUALIZAÇÃO COM PLOTLY

mapa_cores = {
    0: '#A333FF',    # Roxo Elixir (Log Bait)
    1: '#FF3333',    # Vermelho Torre (Beatdown)
    2: '#1A53FF',    # Azul Rei (Ciclo Rápido)
    3: '#FFCC00',    # Dourado Lendário (Champions)
    4: '#2EB82E',    # Verde Veneno/Suporte (Suporte)
    5: '#404040'     # Cinza Estrutura (Construções/Defesa)
}

nomes_cluster = {
    0: 'Log Bait',
    1: 'Beatdown',
    2: 'Ciclo Rápido',
    3: 'Champions/Lendárias',
    4: 'Tank/Suporte',
    5: 'Construções/Defesa '
}

max_pr, min_pr = max(pr.values()), min(pr.values()) ## max e min para normalização posterior
node_traces_list = []
for cluster_id, nome_custom in nomes_cluster.items():
    nodes_in_cluster = [n for n in G.nodes() if clusters[n] == cluster_id]

    trace = go.Scatter(
        x=[pos[n][0] for n in nodes_in_cluster],
        y=[pos[n][1] for n in nodes_in_cluster],
        mode='markers+text',
        name=nome_custom,
        text=nodes_in_cluster,
        textposition="top center",
        hoverinfo='text',
        hovertext=[
            f"<b>{n}</b><br>Arquétipo: {nome_custom}<br>" +
            f"Importância (PR): {pr[n]:.4f}<br>Ponte (BT): {between[n]:.4f}"
            for n in nodes_in_cluster
        ],
        marker=dict(
            size=[((pr[n] - min_pr) / (max_pr - min_pr) * 55 + 15) for n in nodes_in_cluster],
            color=mapa_cores.get(cluster_id),
            line=dict(width=1, color='white'),
            opacity=0.9
        ),
        visible=True # Todos começam visíveis
    )
    node_traces_list.append(trace)

# 2. Edges (Linhas de conexão) - Mantemos em um único grupo ou podemos ocultar no filtro
edge_traces = []
max_w = max([G[u][v]['weight'] for u, v in G.edges()])
for u, v in G.edges():
    x0, y0 = pos[u]; x1, y1 = pos[v]
    w = G[u][v]['weight']
    edge_traces.append(go.Scatter(
        x=[x0, x1, None], y=[y0, y1, None],
        line=dict(width=(w/max_w)*9, color='rgba(150,150,150,0.15)'),
        mode='lines',
        hoverinfo='none',
        showlegend=False,
        visible=True
    ))

# 3. Criar os botões do Menu Lateral
botoes = []

# Botão para mostrar tudo
botoes.append(dict(
    label="Mostrar Todos",
    method="update",
    args=[{"visible": [True] * (len(edge_traces) + len(node_traces_list))}]
))

# Botões para cada cluster
for i in range(len(node_traces_list)):
    # Criamos a lista de visibilidade:
    # 1. Deixamos TODAS as arestas visíveis (True)
    # 2. Deixamos apenas o nó do cluster atual visível (True)
    vis_edges = [True] * len(edge_traces)
    vis_nodes = [False] * len(node_traces_list)
    vis_nodes[i] = True

    vis_total = vis_edges + vis_nodes

    botoes.append(dict(
        label=nomes_cluster[i],
        method="update",
        args=[{"visible": vis_total}]
    ))

# 4. Montar a Figura com o Menu à Direita
fig = go.Figure(data=edge_traces + node_traces_list)

fig.update_layout(
    title=f"Rede de interações Clash Royale (Filtro por Arquétipo)",
    plot_bgcolor='white',
    xaxis=dict(showgrid=False, zeroline=False, showticklabels=False),
    yaxis=dict(showgrid=False, zeroline=False, showticklabels=False),

    # Margem à direita para o menu não sobrepor o gráfico
    margin=dict(r=160, t=100),

       updatemenus=[dict(

        type="dropdown",

        direction="down",

        x=0.1,

        y=1.3,

        showactive=True,

        buttons=botoes,

        bgcolor="rgba(230, 230, 230, 0.8)"

    )]

)



# GERA O ARQUIVO HTML
fig.write_html('clash_analytics_final.html', config={'displayModeBar': False})

# SALVA O CSV
metrics_df['Cluster_Nome'] = metrics_df['Cluster'].map(nomes_cluster)
metrics_df.to_csv('metricas_clash_final.csv', index=False)

print("✅ Dashboard interativo e CSV gerados!")
print("✅ Arquivos gerados com sucesso: HTML (limpo) e CSV (com nomes)!")
```
